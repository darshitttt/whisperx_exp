```python
import os

nisp_dir = '../audioData/NISP-Dataset/'
```

```python
# Creating a new dataframe for the NISP Dataset
aud_fname = []
speaker_id = []
for dir, subdir, files in os.walk(nisp_dir):
    for file in files:
        if file.endswith(".wav"):
            fname = (os.path.join(dir, file))
            aud_fname.append(fname)
            spkr_id = '_'.join(fname.split('/')[-1].split('_')[0:2])
            speaker_id.append(spkr_id)
```

```python
import pandas as pd
import numpy as np

nisp_lab_df = pd.read_table('../audioData/NISP-Dataset/total_spkrinfo.list', sep=" ")
```

```python
(nisp_lab_df[nisp_lab_df['Speaker_ID'] == speaker_id[0]].iloc[0,1])
```

```python
# Finding the age and gender of the speakers
speaker_age = []
speaker_gender = []
for speaker in speaker_id:
    df_row = nisp_lab_df.loc[nisp_lab_df['Speaker_ID'] == speaker]
    speaker_age.append(int(df_row.iloc[0,7]))
    speaker_gender.append(str(df_row.iloc[0,1]))
```

```python
nisp_df = pd.DataFrame({'audio_file': aud_fname, 'speaker_id': speaker_id, 'label': speaker_gender, 'age': speaker_age})
```

```python
nisp_df.to_csv('nisp_genderAge.csv', index=False)
```
