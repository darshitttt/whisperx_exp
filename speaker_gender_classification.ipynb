```python
import pandas as pd
import numpy as np
import model_utils
import os
import librosa
```

```python
# Function to extract features (same as before)
def extract_features(file_path):
    y, sr = librosa.load(file_path)
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr, n_fft=512).T, axis=0)
    mel = np.mean(librosa.feature.melspectrogram(y=y, sr=sr, n_fft=512).T, axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(y=y, sr=sr, n_fft=512).T, axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=y, sr=sr).T, axis=0)
    return np.hstack([mfccs, chroma, mel, contrast, tonnetz])
```

```python
feature_extractor = {
    'feature_extractor_fn': extract_features, 
    'scaler': model_utils.scaler, 
    'label_encoder': model_utils.label_encoder
    }
```

```python
gender_class_ml_models = model_utils.MLModelManager(all_models=True)
combined_dataset = model_utils.Dataset_MLModel('kaggle_voxceleb_combined.csv', feature_extractor)
```

```python
X_train, X_test, y_train, y_test = combined_dataset.get_train_test_split()
eval_dict = gender_class_ml_models.train_classification_MLModels(X_train, y_train, X_test, y_test)
```

```python
gender_class_ml_models.save_classification_MLModels('trained_models/speakerGender_KaggleAndVoxCeleb/')
```

```python
pd.DataFrame(eval_dict).to_csv('trained_models/speakerGender_KaggleAndVoxCeleb/report.csv', index=False)
```
